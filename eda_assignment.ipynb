# Distribution plots

```python
# Plotting distributions
plt.figure(figsize=(14, 10))

# Age Distribution
plt.subplot(2, 2, 1)
data['Age'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')

# Year Distribution
plt.subplot(2, 2, 2)
data['Year'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Year Distribution')
plt.xlabel('Year')
plt.ylabel('Count')



```

```python
# Month Distribution
plt.subplot(2, 2, 3)
data['Month'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Month Distribution')
plt.xlabel('Month')
plt.ylabel('Count')

# Sex Distribution
plt.subplot(2, 2, 4)
data['Sex'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Sex Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Race Distribution
plt.figure(figsize=(10,5))

data['Race'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Race Distribution')
plt.xlabel('Race')
plt.ylabel('Count')
plt.show()

```

```python
# Reason Distribution
plt.subplot(1,2,2)
data['Reason'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Reason Distribution')
plt.xlabel('Reason')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

# Conclusion
Age: The age distribution shows a wide range, with  noticeable concentration around younger people.

Month: July has the highest number of incidents.

Sex: Most of incidents involve males.

Race: Most incidents involve Whites, followed by Blacks and Hispanics.

Reason: Most of the incidents are suicides, followed by homicides.

## Resasons for incident by Race

```python
# Reason for Incident by Race
plt.figure(figsize=(14, 7))
reason_race_df = data.groupby(['Reason', 'Race']).size().unstack().fillna(0)
reason_race_df.plot(kind='bar', stacked=True, edgecolor='black')
plt.title('Reason for Incident by Race')
plt.xlabel('Reason')
plt.ylabel('Count')
plt.show()
```

## Scatter plot


```python
# Scatter Plot: Age vs. Year, colored by Sex
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Year', y='Age', hue='Sex', alpha=0.7)
plt.title('Scatter Plot: Age vs. Year, colored by Sex')
plt.xlabel('Year')
plt.ylabel('Age')
plt.legend(title='Sex')
plt.show()
```

## Age distribution by reason box plot

```python
# Box Plot: Age distribution by Reason
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Reason', y='Age', palette='Set3')
plt.title('Box Plot: Age distribution by Reason')
plt.xlabel('Reason')
plt.ylabel('Age')
plt.show()
```

## Distribution of incidents by place of incident : pie chart

```python
# Pie Chart: Distribution of Incidents by Place of Incident
place_counts = data['Place of incident'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(place_counts, labels=place_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Incidents by Place of Incident')
plt.show()
```

## Distribution by education

```python

education_counts = data['Education'].value_counts()

# Create a mapping from index to education level
education_labels = {i: label for i, label in enumerate(education_counts.index)}

# Create a list of indices for the pie chart
indices = [str(i) for i in range(len(education_counts))]

plt.figure(figsize=(12, 8))
plt.pie(education_counts, labels=indices, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'),
        wedgeprops=dict(width=0.4), pctdistance=0.85)
plt.title('Distribution of Incidents by Education')
plt.gca().add_artist(plt.Circle((0,0),0.7,fc='white'))
plt.show()
```

## Heatmap of incidents by education and reason

```python
# Create a pivot table for the heatmap
heatmap_data = data.pivot_table(index='Education', columns='Reason', aggfunc='size', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Incidents by Education and Reason')
plt.xlabel('Reason')
plt.ylabel('Education Level')
plt.show()
```
